```python
#imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('titanic')
data
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
sns.countplot(x ='survived', data = data)
```

```python
sns.countplot(data=data, x='sex', hue= 'survived')
```

```python
sns.heatmap(data.select_dtypes(exclude=['object','category']).corr(), annot= True, cmap= 'coolwarm', linewidths = 1, linecolor = 'black')
```

From the above corelation matrix, it is clear that 'Fare' and 'Survived' have a positive corelation.<br>
Meaning higher the cost of the ticket, higher is the chance of survival.<br>

```python
sns.regplot(data=data,x='pclass',y='fare')
```

```python
sns.histplot(data,x="fare",bins=15,binwidth=20)
```

```python
sns.histplot(data = data, x = 'fare', hue = 'survived',kde = True);
```

```python
sns.catplot(x='pclass', y='fare', data=data, kind='bar')
```
