Data Fortmating and Normalization


```python
df['Branch'] = df['Branch'].astype('category')
df['City'] = df['City'].astype('category')
df['Customer type'] = df['Customer type'].astype('category')
df['Gender'] = df['Gender'].astype('category')
df['Product line'] = df['Product line'].astype('category')
df['Payment'] = df['Payment'].astype('category')

df.dtypes
```

```python
#before normalization
df.head().select_dtypes(include=['int64', 'float64'])
```

```python
# Selecting numerical columns for normalization
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Applying MinMaxScaler
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Displaying the head of the dataframe to show normalized data
df.head().select_dtypes(include=['int64', 'float64'])
```

```python
# Label Encoding
df['Customer type Label Encoded'] = LabelEncoder().fit_transform(df['Customer type'])
df['Payment Label Encoded'] = LabelEncoder().fit_transform(df['Payment'])

# Dummy Encoding
df_customer_dummy = pd.get_dummies(df['Customer type'], prefix='Customer',drop_first=True)
df_payment_dummy = pd.get_dummies(df['Payment'], prefix='Payment',drop_first=True)

# One-Hot Encoding
onehot_encoder = OneHotEncoder()
customer_onehot = onehot_encoder.fit_transform(df[['Customer type']])
customer_onehot = pd.DataFrame(customer_onehot.toarray(), columns=[f'Customer_{cat}' for cat in onehot_encoder.categories_[0]])
payment_onehot = onehot_encoder.fit_transform(df[['Payment']])
payment_onehot = pd.DataFrame(payment_onehot.toarray(), columns=[f'Payment_{cat}' for cat in onehot_encoder.categories_[0]])

print('Label Encoding for Customer type and Payment:')
print(df[['Invoice ID','Customer type', 'Customer type Label Encoded']].head())
print(df[['Invoice ID','Payment', 'Payment Label Encoded']].head())

print('\
Dummy Encoding for Customer type and Payment:')
print(df_customer_dummy.head())
print(df_payment_dummy.head())
print('\
One-Hot Encoding for Customer type and Payment:')
print(customer_onehot.head())
print(payment_onehot.head())
```
