```python
#imports 
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data, columns = iris.feature_names)
data['label'] = iris.target
data.sample(5)
```

```python
data.info()
```

## Histograms

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
for i, col in enumerate(data.columns):
    if i == 4: break
    plt.subplot(2, 2, i+1)
    sns.histplot(data[col], kde=True, color='skyblue', edgecolor='black')
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python

# Create box plots for each feature
plt.figure(figsize=(12, 8))
sns.boxplot(data=data)
plt.title('Box Plots for Iris Dataset Features')
plt.show()
```

```python
sns.histplot(data = data, x = 'sepal length (cm)', kde= True);
```

```python
sns.histplot(data = data, x = 'sepal width (cm)', kde= True, color = "orange");
```

```python
sns.histplot(data = data, x = 'petal length (cm)', kde= True, color = "green");
```

```python
sns.histplot(data = data, x = 'petal width (cm)', kde= True, color = "red");
```

## Box Plots

```python
figure = plt.figure(figsize = (12,8))
sns.boxplot(data= data)
plt.show()
```

There are some outliers present in the 'sepal width (cm)' attribute. <br>

```python
from matplotlib.cbook import boxplot_stats
stats = boxplot_stats(data['sepal width (cm)'])
stats
```
